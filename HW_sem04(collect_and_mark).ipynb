```python
# Импорт необходимых библиотек

from lxml import html
import requests
import json
import pandas as pd
import re
```

Будем работать с сайтом "Всероссийской федерации плавания" и таблицей "Тренеры и специалисты". Таблица настолько "кривая", что я просто не смог пройти мимо.

```python
# Сохраним адрес сайта в отдельную переменную
url = 'https://new.russwimming.ru/data/list/treneri_specialisti.html'
```

```python
# Создадим переменную запроса к сайту с использованием агента пользователя
resp = requests.get(url=url, headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'})

# Получим код статуса ответа сайта
resp.status_code
```

```python
tree = html.fromstring(resp.content)
```

Поскольку таблица на сайте разделена на секции "Плавание. Мужчины", "Плавание. Женщины" и "Плавание на открытой воде", то, как вариант, можно сделать парсинг для каждой секции. Возможно есть специальный прием для обработки таких таблиц, как единого целого, но он мне пока неизвестен.
